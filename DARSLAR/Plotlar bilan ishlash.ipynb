**`subplot usuli`**

```python
rasm = mimg.imread("RASMLAR/Tashkent.jpg")
plt.imshow(rasm)
red_chanel = rasm[:,:,0]
green_chanel = rasm[:,:,1]
blue_chanel = rasm[:,:,2]
```

```python
%matplotlib inline
rasm = mimg.imread("RASMLAR/Tashkent.jpg")
plt.title("Haqiqiy rasm")
plt.imshow(rasm)

red_chanel = rasm[:,:,0]
green_chanel = rasm[:,:,1]
blue_chanel = rasm[:,:,2]

plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
plt.title("Red")
plt.imshow(red_chanel)

plt.subplot(1,3,2)
plt.title("Green")
plt.imshow(green_chanel)

plt.subplot(1,3,3)
plt.title("Blue")
plt.imshow(blue_chanel)

plt.show()
```

#### Subplots usuli

```python
rasm = mimg.imread("RASMLAR/Tashkent.jpg")
plt.imshow(rasm)
red_chanel = rasm[:,:,0]
green_chanel = rasm[:,:,1]
blue_chanel = rasm[:,:,2]
```

```python
figure, (plt1,plt2,plt3) = plt.subplots(1,3,figsize=(15,8))

plt1.set_title("Red chanel")
plt1.set_xlabel("Eni")
plt1.set_ylabel("Balandligi")
plt1.imshow(red_chanel)

plt2.set_title("Green chanel")
plt2.set_xlabel("Eni")
plt2.set_ylabel("Balandligi")
plt2.imshow(green_chanel)

plt3.set_title("Blue chanel")
plt3.set_xlabel("Eni")
plt3.set_ylabel("Balandligi")
plt3.imshow(blue_chanel)

plt.show()
```

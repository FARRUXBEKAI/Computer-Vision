```python
import cv2
import numpy as np
```

```python
# Har hil ranglar
rasm1 = np.full((150,150,3),0,dtype=np.uint8)
rasm2 = np.full((150,150,3),50,dtype=np.uint8)
rasm3 = np.full((150,150,3),100,dtype=np.uint8)
rasm4 = np.full((150,150,3),150,dtype=np.uint8)
rasm5 = np.full((150,150,3),200,dtype=np.uint8)
rasm6 = np.full((150,150,3),250,dtype=np.uint8)

x = np.hstack((rasm1,rasm2))
y = np.hstack((rasm3,rasm4))
m = np.hstack((rasm5,rasm6))
z = np.vstack((x,y,m))
z = np.hstack((z,z))
cv2.imshow("Rasm",z)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Ism: FARRUH oq qora
qora = np.full((20,20,3),0,dtype=np.uint8)
oq = np.full((20,20,3),255,dtype=np.uint8)
x1 = np.hstack((qora,qora,qora))
x2 = np.hstack((qora,oq,oq))
x3 = np.hstack((qora,oq,qora))
x4 = np.hstack((qora,qora,oq))

f = np.vstack((x1,x2,x1,x2,x2))
a = np.vstack((x1,x3,x1,x3,x3))
r = np.vstack((x1,x3,x1,x4,x3))
u = np.vstack((x3,x3,x3,x3,x1))
h = np.vstack((x3,x3,x1,x3,x3))
o = np.vstack((oq,oq,oq,oq,oq))

ism = np.hstack((f,o,a,o,r,o,r,o,u,o,h))

cv2.imshow("ISM",ism)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Ism: FARRUH rangli
qora = np.random.randint(20,250,size=(20,20,3),dtype=np.uint8)
oq = np.full((20,20,3),255,dtype=np.uint8)
x1 = np.hstack((qora,qora,qora))
x2 = np.hstack((qora,oq,oq))
x3 = np.hstack((qora,oq,qora))
x4 = np.hstack((qora,qora,oq))

f = np.vstack((x1,x2,x1,x2,x2))
a = np.vstack((x1,x3,x1,x3,x3))
r = np.vstack((x1,x3,x1,x4,x3))
u = np.vstack((x3,x3,x3,x3,x1))
h = np.vstack((x3,x3,x1,x3,x3))
o = np.vstack((oq,oq,oq,oq,oq))

ism = np.hstack((f,o,a,o,r,o,r,o,u,o,h))

cv2.imshow("ISM",ism)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Rasm kanallarini o'zgartirish

```python
rasm_opencv = cv2.imread("RASMLAR/Masjid.jpg")
rasm_opencv = cv2.cvtColor(rasm_opencv,cv2.COLOR_BGR2RGB) # BGR dan RGB ga o'tqazish
plt.imshow(rasm_opencv)
plt.show()
```

```python
rasm_opencv2 = cv2.imread("RASMLAR/python3.jpg")
rasm_opencv2 = cv2.cvtColor(rasm_opencv2,cv2.COLOR_BGR2RGB) # BGR dan RGB ga o'tqazish
plt.imshow(rasm_opencv2)
plt.show()
```

```python
rasm_opencv2 = cv2.imread("RASMLAR/python3.jpg")
rasm_opencv2 = cv2.cvtColor(rasm_opencv2,cv2.COLOR_BGR2GRAY) # BGR dan GRAY ga o'tqazish yani qora oq ga
plt.imshow(rasm_opencv2,cmap="gray") #oqqorada o'qishini aytib qo'yish kerak unda birinchi kelgan kanalda o'qib ketaveradi
plt.show()
```

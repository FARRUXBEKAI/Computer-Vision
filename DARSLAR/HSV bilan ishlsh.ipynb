# HSV - hue saturation value
* HSV da raqamlar 360 gacha bo'lgan lekin uni minimallashtirib 0-180 oralig'iga olib kelingan.
* Hue-qaysi rangligi,saturation va value rangning qalinligi 

#### HUE ( 0 - 180 )
* `0-60 300-360 -> Red`
* `60-120 120-180 -> Green`
* `180-240 240-300 -> Blue`

* `HSV = [0-180,0-255,0-255]` `[rang,rangdan qancha darjada qo'shilganligi,yorqinlik darajasi]`

* `0 - 30 150 - 180 -> Red`
* `30 - 90 -> Green`
* `90 - 150 -> Blue`

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

**`Oddiy RGB usulida ajratib olish ozgina kamchilik bilan`**

```python
rasm = cv2.imread("RASMLAR/car_green_screen2.jpg")
rasm = cv2.cvtColor(rasm,cv2.COLOR_BGR2RGB)

range_1 = np.array([0,40,0])
range_2 = np.array([150,255,150])

maska = cv2.inRange(rasm,range_1,range_2)
rasm_copy = rasm.copy()
rasm_copy[maska!=0] = 0

plt.figure(figsize=(12,8))

plt.subplot(131)
plt.imshow(rasm)

plt.subplot(132)
plt.imshow(maska,cmap="gray")

plt.subplot(133)
plt.imshow(rasm_copy)
```

**`HSV dan foydalanib aniqroq ajratib olish`**

```python
rasm = cv2.imread("RASMLAR/car_green_screen2.jpg")
rasm_rgb = cv2.cvtColor(rasm,cv2.COLOR_BGR2RGB)
rasm_hsv = cv2.cvtColor(rasm,cv2.COLOR_BGR2HSV)

range_1 = np.array([40,0,0])
range_2 = np.array([100,250,250])

maska = cv2.inRange(rasm_hsv,range_1,range_2)
rasm_rgb[maska!=0] = 0

plt.figure(figsize=(12,8))

plt.subplot(131)
plt.imshow(rasm_hsv)

plt.subplot(132)
plt.imshow(maska,cmap="gray")

plt.subplot(133)
plt.imshow(rasm_rgb)
```

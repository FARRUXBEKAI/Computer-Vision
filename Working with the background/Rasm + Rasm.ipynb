```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

**`Asfaltga mashina qo'yish`**

```python
car = cv2.imread("RASMLAR/car2.jpg")
car = cv2.cvtColor(car,cv2.COLOR_BGR2RGB)
plt.imshow(car)
```

```python
car_copy = np.copy(car)
range_1 = np.array([0,220,0])
range_2 = np.array([200,255,200])
maska = cv2.inRange(car_copy,range_1,range_2)
plt.imshow(maska,cmap="gray")
```

```python
car_copy[maska!=0] = [0,0,0]
plt.imshow(car_copy)
```

```python
background = cv2.imread("RASMLAR/asfalt2.jpg")
background = cv2.cvtColor(background,cv2.COLOR_BGR2RGB)
background = cv2.resize(background,(300,168))
plt.imshow(background)
```

```python
background_copy = background.copy()
background_copy[maska==0] = [0,0,0]
plt.imshow(background_copy)
```

```python
new_photo = car_copy + background_copy
plt.imshow(new_photo)
```

```python
plt.figure(figsize=(12,8))

plt.subplot(141)
plt.title("Haqiqiy")
plt.imshow(car)

plt.subplot(142)
plt.title("Maska")
plt.imshow(maska,cmap="gray")

plt.subplot(143)
plt.title("Background")
plt.imshow(background)

plt.subplot(144)
plt.title("New photo")
plt.imshow(new_photo)
```
